```python
%matplotlib inline

from matplotlib import pyplot
#import matplotlib.pyplot as plt
import pylab
from mpl_toolkits.mplot3d import Axes3D
import random
fig = pylab.figure()
ax=Axes3D(fig)

x1=[]
y1=[]
z1=[]
x1d = 3.
y1d = 3.
z1d = 3.

x2=[]
y2=[]
z2=[]
x2d = 3.
y2d = 3.
z2d = -3.

x3=[]
y3=[]
z3=[]
x3d = 3.
y3d = -3.
z3d = 3.


for i in range(0,500):
    x1.append(random.gauss(0,1)+x1d)
    y1.append(random.gauss(0,1)+y1d)
    z1.append(random.gauss(0,1)+z1d)
    
    x2.append(random.gauss(0,1)+x2d)
    y2.append(random.gauss(0,1)+y2d)
    z2.append(random.gauss(0,1)+z2d)
    
    x3.append(random.gauss(0,1)+x3d)
    y3.append(random.gauss(0,1)+y3d)
    z3.append(random.gauss(0,1)+z3d)

ax.scatter(x1,y1,z1)
ax.scatter(x2,y2,z2)
ax.scatter(x3,y3,z3)

for i in range(0,2):
    ax.azim = i*2
    ax.elev = -45
    ax.text(3, 3, 3, "apple", color='red')
    filename=str(i).zfill(2)+'test.png'
    print(filename,ax.azim)
    pyplot.savefig(filename)
```
